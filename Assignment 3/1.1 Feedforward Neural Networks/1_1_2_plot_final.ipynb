```python
import os
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
fileName = 'temp_0.0001000.npy'
data = np.load(fileName, encoding = 'latin1')
```

```python
trainLoss = data.item().get('trainLoss')
validLoss = data.item().get('validLoss')
testLoss = data.item().get('testLoss')

trainAcc = data.item().get('trainAcc')
validAcc = data.item().get('validAcc')
testAcc = data.item().get('testAcc')
```

```python
xRange = np.arange(len(trainLoss))

fig = plt.figure()

plt.plot(xRange, trainLoss, '-', label='train loss')
plt.plot(xRange, validLoss, '-', label='valid loss')
plt.plot(xRange, testLoss, '-', label='test loss')
plt.plot([0, len(trainLoss)], [0.31, 0.31], color='k', linestyle='-', linewidth=1)

plt.ylim((0.0, 0.8))
plt.xlim((-1, 700))

plt.xlabel('Number of Epochs')
plt.ylabel('Cross Entropy Loss')
plt.legend()
plt.show()
```

```python
xRange = np.arange(len(trainAcc))

fig = plt.figure()

plt.plot(xRange, trainAcc, '-', label='train accuracy')
plt.plot(xRange, validAcc, '-', label='valid accuracy')
plt.plot(xRange, testAcc, '-', label='test accuracy')
#plt.plot([0, len(trainAcc)], [0.31, 0.31], color='k', linestyle='-', linewidth=1)

plt.ylim((0.5, 1.0))
plt.xlim((-1, 200))

plt.xlabel('Number of Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
