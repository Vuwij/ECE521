```python
import os
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = np.load('temp_500.0000000.npy', encoding = 'latin1')
data1.item().get('validAcc')

data2 = np.load('double_temp.npy', encoding = 'latin1')
trainAcc2 = data2.item().get('trainAcc')
vaildAcc2 = data2.item().get('validAcc')
testAcc2 = data2.item().get('testAcc')
```

```python
xRange = np.arange(len(trainAcc2))

fig = plt.figure()

plt.plot(xRange, trainAcc2, '-', label='train accuracy')
plt.plot(xRange, vaildAcc2, '-', label='valid accuracy')
plt.plot(xRange, testAcc2, '-', label='test accuracy')

#plt.ylim((0.5, 1))
#plt.xlim((-1, 300))
plt.xlabel('Number of Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
