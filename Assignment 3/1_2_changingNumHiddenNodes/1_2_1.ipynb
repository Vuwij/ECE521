```python
import os
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
files = os.listdir(os.curdir)
print(files)

found_list = []

for file in files:
    if "temp_" in file:
        found_list.append(file)

print(found_list)
```

```python
dataList = []
for fileName in found_list:
    dataList.append(np.load(fileName, encoding = 'latin1'))
    
trainLossList = []

for data in dataList:
    trainLossList.append(data.item().get('validAcc'))
```

```python
xRange = np.arange(len(trainLossList[0]))

fig = plt.figure()

for i in range(len(trainLossList)):
    plt.plot(xRange, trainLossList[i], '-', label='WD Coef %s' % (found_list[i]))
plt.ylim((0.5, 1))
plt.xlim((-1, 300))
plt.xlabel('Number of Epochs')
plt.ylabel('validAcc')
plt.legend()
plt.show()
```
